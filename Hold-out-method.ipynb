```python
# 1 - hold out Method
import pandas as pd 
data = pd.read_csv('diabetics.csv') 
# data ===> dataframe
```

```python
# x = all columns - end column
x = data.drop("outcome",axis=1)
# y = end column
y = data['outcome']
# كدا قسمت الداتا ل x , y ( feature , label)
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test=train_test_split(x,y,test_size=0.3)  
# لازم بالترتيب                                         feature label  النسبة الي هقسم بيها(30و70)
```

```python
data.head() 
# بيظهر اول 5 ريكورد
```

```python
data.shape
#الداتا فيها كام  صف و كام عموووود 
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 50)
classifier.fit(x_train, y_train)
#الي هتخليه يعرف يتعلم 
```

```python
predictions = classifier.predict(x_test)
#الي هسجل فيه الاجابة 
```

```python
from sklearn.metrics import confusion_matrix
matrix = confusion_matrix(predictions,y_test)
#الشخص الي هيصحح و يقارن 
print('confusion matrix ',matrix)
```

```python
from sklearn import metrics
acc = metrics.accuracy_score(y_test, predictions)
print("Accuracy :",acc)
## accuracy = (true Positive + True Negative) / count of y_test ==> (110+113)/(110+71+52+113) = 0.644508
```

```python
pre = metrics.precision_score(y_test, predictions)
print("precition:",pre)
#precition = true Positive / (true Positive + false Positive) ==> (110)/ (110+52) =0.68484848
```

```python
recall = metrics.recall_score(y_test, predictions)
print("recall:",recall)
# recall = true Positive / (true Positive + false negative) ==>  (110)/(110+71) =0.61413043
```
